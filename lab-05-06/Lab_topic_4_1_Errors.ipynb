## 2. try/except/finally
**Perform cleanup actions, whether exceptions occur or not**

Частина 2: try/except/finally
Продемонструвати, що finally виконується завжди

Попросити користувача ввести новий шлях (наприклад, через input())

ЗАВДАННЯ:
Модифікуй код так, щоб після finally файл відкривався у новій директорії, якщо вона правильна.

```python
try:
    filename = "/home/gf/src/Python/Python-in-the-lap/Bk/F64ac_0.03_time_200.dat"
    data = -np.loadtxt(filename)
except OSError:
    print("%s not found" % filename)
finally:
    new_input = input("Give me a new directory: ")

if os.path.isdir(new_input):
    new_filename = os.path.join(new_input, "F64ac_0.03_time_200.dat")
    if os.path.isfile(new_filename):
        data = -np.loadtxt(new_filename)
        print("File opened successfully:", new_filename)
    else:
        print("File not found in the new directory:", new_filename)
else:
    print("Invalid directory:", new_input)
```

## 3. raise
**Trigger an exception manually**

Частина 3: raise
Показати, як штучно згенерувати виняток

Створити перевірку, яка викликає помилку вручну, якщо значення не відповідає умові

ЗАВДАННЯ:
Створи свою функцію, яка викидає ValueError, якщо число більше 100.

### Sometime we can create an error on purpose

```python
def check_number(value):
    if value > 100:
        raise ValueError("The number must not be greater than 100")
    return value

print(check_number(33))
check_number(101)
```

```python
try:
    q = 1 / 0
    print(q)
except ZeroDivisionError:
    print("Check the division")
    q = 1
```

## 4. assert
**Conditional trigger an exception in the code**

Частина 4: assert
Продемонструвати assert як засіб перевірки

Пояснити, коли доцільно використовувати (при розробці, не у продакшні)

ЗАВДАННЯ:
Реалізуй функцію, яка приймає число і перевіряє, що воно додатне, з допомогою assert.

```python
def check_positive(x):
    assert x > 0, "x must be positive"
    return x
```

```python
check_positive(5)
```

```python
check_positive(-1)
```

## 5. with/as
**Implement context managers**

Частина 5: with/as
Показати, як працює відкриття файлу через with

Пояснити перевагу: автоматичне закриття, без f.close()

ЗАВДАННЯ:
Відкрий будь-який .txt або .md файл у своїй директорії, виведи перші 5 рядків.

```python
from pathlib import Path

candidates = list(Path('.').glob('*.md')) + list(Path('.').glob('*.txt'))
if not candidates:
    sample_file = Path('sample_notes.txt')
    sample_file.write_text('Line 1\nLine 2\nLine 3\nLine 4\nLine 5\nLine 6\n')
    candidates = [sample_file]

file_to_open = candidates[0]
print('Using file:', file_to_open)
with open(file_to_open) as f:
    data = f.readlines()
```

```python
for line in data[:5]:
    print(line.rstrip())
```
